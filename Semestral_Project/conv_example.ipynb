```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from scipy import ndimage
```

```python
img = cv2.cvtColor(cv2.imread('./among_us_pixel_art_by_nimowerytheking_deticwb-fullview.jpeg'), cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap='gray')
```

```python
kernel1 = np.array([[0.25, 0.5, 0.25], 
                [0.0, 0.0, 0.0], 
                [-0.25, -0.5, -0.25]])
```

```python
plt.imshow(ndimage.convolve(img, kernel1, mode='nearest', cval=1.0), cmap='gray')
```

```python
kernel2 = np.array([[0.25, 0.0, -0.25], 
                [0.5, 0.0, -0.5], 
                [0.25, 0.0, -0.25]])
```

```python
plt.imshow(ndimage.convolve(img, kernel2, mode='nearest', cval=1.0), cmap='gray')
```
