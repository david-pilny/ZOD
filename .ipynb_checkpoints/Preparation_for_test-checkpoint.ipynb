```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## [2] Kytka - Word 1 

```python
img_kytka = cv2.cvtColor(cv2.imread('./OneDrive_1_11-10-2022/kytka256.jpg'), cv2.COLOR_BGR2RGB)
gray_kytka = np.stack((cv2.cvtColor(cv2.imread('./OneDrive_1_11-10-2022/kytka256.jpg'), cv2.COLOR_BGR2GRAY),)*3, axis=-1)

background = np.ones((800,1000, 3), dtype=int)
background[0:799, 0:999, :] = [240, 240, 240]

background[50:len(img_kytka)+50, 100:len(img_kytka)+100] = img_kytka
background[500:len(gray_kytka)+500, 650:len(gray_kytka)+650] = gray_kytka
plt.imshow(background)
```

## [6] Vybarven kvtina - Word 1

## [2] Kytka v prostedku - Word 2 

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

img = plt.imread('./OneDrive_1_11-10-2022/kytka256.jpg', format='jpg')

fig, ax = plt.subplots()

imagebox = OffsetImage(img, zoom=0.4)
imagebox.image.axes = ax

ab = AnnotationBbox(imagebox, (0.5, 0.5), xycoords='axes fraction',
                    bboxprops={'lw':0})

ax.add_artist(ab)

plt.show()
```

## [3] Gradient - Word 2

## [3] Kytkostn - Word 2

```python
img_kytka = cv2.cvtColor(cv2.imread('./OneDrive_1_11-10-2022/kytka256.jpg'), cv2.COLOR_BGR2RGB)
gradient = np.ones((256,256, 3), dtype=int)

for x in range(0, 256):
    for y in range(0, 256):
        gradient[x][y] = y

blended = cv2.addWeighted(img_kytka, 0.5, np.uint8(gradient), 0.7, 0.5)
plt.imshow(blended)
```
